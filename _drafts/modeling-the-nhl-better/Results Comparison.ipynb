```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plot_parameter_estimate(lv_df, iv_df, param):
    plt.figure(figsize=(10, 6))
    plt.title('Estimates for: ' + param)
    plt.plot(lv_df['day'], lv_df[param], color='blue')
    plt.plot(iv_df['day'], iv_df[param+'_μ'], color='red')
    upper1sd = iv_df[param+'_μ'] + iv_df[param+'_σ']
    lower1sd = iv_df[param+'_μ'] - iv_df[param+'_σ']
    upper2sd = iv_df[param+'_μ'] + 2 * iv_df[param+'_σ']
    lower2sd = iv_df[param+'_μ'] - 2 * iv_df[param+'_σ']
    plt.fill_between(iv_df['day'], upper2sd, lower2sd, color='red', alpha=0.2)
    plt.fill_between(iv_df['day'], upper1sd, lower1sd, color='red', alpha=0.2)
    plt.show()

def plot_multi_parameter_estimate(lv_df, iv_df, param_list, y_lim=(-0.6, 0.6), grid_lines=0.10):
    imgsize = 4
    figsize = (15,15)
    rows = int(np.ceil(np.sqrt(len(param_list))))
    fig, axs = plt.subplots(rows, rows, figsize=figsize)
    ax = axs.flatten()
    for i in range(len(param_list)):
        param = param_list[i]
        ax[i].set_title('Estimates for: ' + param)
        ax[i].plot(lv_df['day'], lv_df[param], color='blue')
        ax[i].plot(iv_df['day'], iv_df[param+'_μ'], color='red')
        upper1sd = iv_df[param+'_μ'] + iv_df[param+'_σ']
        lower1sd = iv_df[param+'_μ'] - iv_df[param+'_σ']
        upper2sd = iv_df[param+'_μ'] + 2 * iv_df[param+'_σ']
        lower2sd = iv_df[param+'_μ'] - 2 * iv_df[param+'_σ']
        ax[i].fill_between(iv_df['day'], upper2sd, lower2sd, color='red', alpha=0.2)
        ax[i].fill_between(iv_df['day'], upper1sd, lower1sd, color='red', alpha=0.2)
        for y in np.arange(y_lim[0] + grid_lines, y_lim[1], grid_lines):
            ax[i].hlines(y, 1, n_days, colors='k', linestyles='dotted', alpha=0.4)
        ax[i].set_ylim(y_lim[0], y_lim[1])
        
    fig.tight_layout()
    plt.show()
```

### MSE against known latent variables

```python
n_days = 400
comparison_dates = list(range(170, 400))
```

```python
param_list = ['i','h'] + ['o{}'.format(i) for i in range(32)] + ['d{}'.format(i) for i in range(32)]

def compute_mse_mae(model_type):
    iv_df = iv_dfs[model_type]
    n = 0
    sse = 0
    serr = 0
    for i in range(len(param_list)):
        param = param_list[i]
        #lvs = lv_df[lv_df['day'].isin(comparison_dates)].sort_values('day')[param]
        #ivs = iv_df[iv_df['day'].isin(comparison_dates)].sort_values('day')[param+'_μ']
        vs = lv_df[['day', param]].merge(iv_df[['day', param+'_μ']], how='left', on=['day'])
        vs = vs.dropna()
        serr += np.sum(np.sqrt((vs[param] - vs[param+'_μ'])**2))
        sse += np.sum(((vs[param] - vs[param+'_μ'])**2))
        n += vs.shape[0]
        
    mse = sse/n
    mae = serr/n
    return mse, mae
```

```python
for mt in iv_dfs.keys():
    mse, mae = compute_mse_mae(mt)
    print(f'Dataset: {mt}, MSE: {mse}, MAE: {mae}')
```
